# **Lectura de archivos excel**
- Ejecutar celda para ingresar el nombre del archivo a leer.

```python
#@title
try:
  input_excel = input("Inserte unicamente el nombre del archivo excel a leer: ") + '.xlsx'
  df = pd.read_excel('/content/drive/MyDrive/Programa_QR/Import/' + input_excel)
  print("Archivo {0} conectado.".format(input_excel))

except:
  print("Ha ocurrido un error, porfavor verifique que haya ingresado correctamente el nombre del archivo a leer.")
```

# **Configuración del código QR**
- Ejecutar celda para configurar el generador de QR

```python
#@title
try:
  qr = qrcode.QRCode(
      version=None,
      error_correction=qrcode.constants.ERROR_CORRECT_L,
      box_size=60,
      border=1
  )
  print("Se ha configurado exitosamente.")

except:
  print("Ha ocurrido un error. Porfavor, verifique que haya ejecutado las celdas en orden.")
```

# **Generación de códigos QR**
- Ejecutar celda para generar imágenes de código QR.
- Se almacenaran en la carpeta "/Images"
- Demora aprox. 3'.

```python
#@title
count = 0
Imagenes_QR = []

try:
  for i in range(len(df)):
      qr.clear()
      qr.add_data(df.loc[i, 'LINK'])
      qr.make(fit=True)
      img = qr.make_image(fill_color="#000000", back_color="#ffffff")
      img = img.resize((62, 62))
      count += 1
      img_name = "Imagen" + str(count) + ".png"
      img.save('/content/drive/MyDrive/Programa_QR/Images/'+ img_name)
      Imagenes_QR.append(img_name)
  print("Creación de códigos QR exitosa.")

except:
  print("Ha ocurrido un error. Porfavor, verifique que haya ejecutado las celdas en orden.")
```

# **Exportación de excel con códigos QR**

- Ejecutar celda para ingresar el nombre del archivo a exportar.

```python
#@title
output_excel = ""

try:
  output_excel = input("Ingresa el nombre del archivo a exportar: ") + ".xlsx"
  print("El archivo a exportar es: {0}".format(output_excel))

except:
  print("Porfavor, ingrese un nombre para el archivo a exportar.")
```

- Ejecutar celda para exportar el archivo con las imágenes QR en la carpeta "/Export".

```python
#@title
wb = load_workbook('/content/drive/MyDrive/Programa_QR/Import/' + input_excel)

ws = wb['Hoja1']

ws.insert_cols(2)

ws['B1'] = "IMAGEN QR"
B1 = ws['B1']
B1.font = Font(bold=True)

count_rows = 1

try:
  for i in range(len(Imagenes_QR)):
    count_rows += 1
    img = Image('/content/drive/MyDrive/Programa_QR/Images/' + Imagenes_QR[i])
    ws.add_image(img, 'B' + str(count_rows))
    count_rows = int(count_rows)

  wb.save('/content/drive/MyDrive/Programa_QR/Export/'+ output_excel)
  print("Exportación exitosa del archivo {0}.".format(output_excel))

except:
  print("Ha ocurrido un error. Porfavor, verifique que haya ejecutado las celdas en orden")
```

# **VACIAR CARPETA "/Images"** ❗❗❗
- Ejecutar celda para borrar todas las imágenes antes de generar otra tirada de códigos QR, para evitar conflictos.

```python
#@title
import os

archivos = os.listdir('/content/drive/MyDrive/Programa_QR/Images/')
try:
  input_remove = input("Desea borrar todas las imágenes de código QR de la carpeta '/Images'? Ingrese Y o N: ")
  if input_remove.upper() == "Y":
    if len(archivos) > 0:
      for i in archivos:
        os.remove('/content/drive/MyDrive/Programa_QR/Images/' + i)
      print("Se removieron todas las imágenes de la carpeta 'Images' exitosamente.")

    else:
      print("No se encontraron imágenes, carpeta vacía.")

  elif input_remove.upper() == "N":
    print("No se borró ninguna imagen de código QR.")

  else:
    print("Porfavor, ejecute nuevamente la celda e ingrese una opción válida. Y o N")

except:
  print("Porfavor, ejecute nuevamente la celda e ingrese una opción válida. Y o N")
```
